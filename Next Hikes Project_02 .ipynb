```python
# Q(Quartile)1 = 25 % ,Q2(Median) = 50 % ,Q3(75%)
```

```python
# To find outlier indexes
q1 = finel_clean_dataset['Total count'].quantile(.25)
q3 = finel_clean_dataset['Total count'].quantile(.75)
iqr = q3-q1
min_r = q1-1.5*iqr
max_r = q3+1.5*iqr
outliers = finel_clean_dataset[(finel_clean_dataset['Total count']<min_r) | (finel_clean_dataset['Total count']>max_r)]
outliers
```

```python
outliers.shape
```

#### Remove outliers

```python
finel_clean_dataset_no_outlier = finel_clean_dataset.drop(outliers.index)
finel_clean_dataset_no_outlier
```
