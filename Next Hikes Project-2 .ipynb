```python
# Calculate the IQR for the specified column 
Q1 = Finel_clean_dataset['cnt'].quantile(0.25)
Q3 = Finel_clean_dataset['cnt'].quantile(0.75)
IQR = Q3-Q1
```

```python
# Define the minimum (min) and maximum(max) bonds for outliers
min_bonds = Q1-1.5*IQR
max_bonds = Q3+1.5*IQR
```

```python
# Find outliers in the specified column 
outliers = Finel_clean_dataset[(Finel_clean_dataset['cnt'] <min_bonds) |(Finel_clean_dataset['cnt'] > max_bonds)]
print(f"outliers in '{'cnt'}':")
print(outliers)
```
